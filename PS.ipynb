```python
from elasticsearch import Elasticsearch

# Spécifiez l'URL de votre cluster Elasticsearch
es = Elasticsearch(['http://localhost:9200'])

# Nom de votre index
index_name = 'nom_de_votre_index'

# Vérifiez si l'index existe déjà
if not es.indices.exists(index=index_name):
    # Définissez le mappage du document (adapté à votre cas)
    mapping = {
        'properties': {
            'Nom': {'type': 'text'},
            'Catégorie': {'type': 'text'},
            'ChiffreAffaire': {'type': 'float'},
            'Année': {'type': 'integer'},
            'Tendance': {'type': 'text'},
            'Génération': {'type': 'text'},
            'Pourcentage_achat': {'type': 'integer'},
            'PaysOrigine': {'type': 'text'},
            'SiteWebPrincipal': {'type': 'text'},
            'LanguePrincipale': {'type': 'text'},
           
        }
    }

    # Créez l'index avec le mappage spécifié
    es.indices.create(index=index_name, body={'mappings': mapping})
    print(f"L'index '{index_name}' a été créé avec succès.")
else:
    print(f"L'index '{index_name}' existe déjà.")
```
